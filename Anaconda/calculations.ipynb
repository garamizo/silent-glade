# Selecting resistor in series to LDR
## LDR Resistance

$R_{LDR}^{day} = 1.5 k \Omega$

$R_{LDR}^{night} = 10.0 k \Omega$

Using voltage divider to trigger MOSFET:

$R_{gate} = V_{cc} \frac{R_{LDR}}{R_{LDR}+R}$

## Series Resistor

```python
import numpy as np
import matplotlib.pyplot as plt

R = np.arange(100, 10e3, 100)

Vday = 3.2*1.5e3/(1.5e3 + R)
Vnight = 3.2*10e3/(10e3 + R)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(R/1e3, Vday, label='day')
ax.plot(R/1e3, Vnight, label='night')
ax.legend()
ax.grid()
ax.set_ylim([0, 3.2])

plt.ylabel('$V_{gate}\ [V]$')
plt.xlabel('$R\ [k\Omega]$')
plt.show()
```
